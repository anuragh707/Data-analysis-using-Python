```python
#get the salary from the user , if the is more or equal to 50000 apply 10% reduction , if the salary is < 50000 apply %5 , if the salary is more than 70000 apply 12% tax and if salary exceeds
salary=int(input("Enter your sarlary="))
if(salary<50000):
  final_salary=salary-salary*(0.05)
  print("final salary =",final_salary)
elif(salary<=70000):
  final_salary=salary-salary*(0.1)
  print("final salary =",final_salary)
elif(salary<=100000):
  final_salary=salary-salary*(0.12)
  print("final salary =",final_salary)
else:
  print("You are promoted")
```
